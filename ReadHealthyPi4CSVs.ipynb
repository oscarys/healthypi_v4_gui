```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Filename
FILE = 'hpi4_20260501_155519.csv'
```

```python
# Read as a dataframe
hpi = pd.read_csv(FILE)
# Display column names
hpi.columns
```

```python
# Plot a short interval of ECG, raw and filtered
plt.plot(hpi['ecg'], label='original')
plt.plot(hpi['ecg_fir_filter'], label='filtrada')
plt.xlim([0, 625])
plt.legend();
```

```python
# Plot scaled and mean-removed PPGs
plt.plot((hpi['ppg_ir']-np.mean(hpi['ppg_ir']))/1000, label='IR')
plt.plot((hpi['ppg_red']-np.mean(hpi['ppg_red']))/1000, label='Red')
plt.xlim([0, 1000])
plt.legend();
```
